## C. Konversi susunan kolom

#### 1. Mengkonversi susunan kolom _Confirmed_, _Cases_, dan _Recovered_ agar hanya menjadi 1 kolom

```python
df = df.melt(id_vars=['SNo', 'ObservationDate', 'Province/State', 'Country/Region', 'Last Update'], 
        var_name="Cases", 
        value_name="Counted")
df
```

#### 2. Menyimpan dataframe ke file csv

```python
df.to_csv('covid_19_melted.csv', index=False)
```
